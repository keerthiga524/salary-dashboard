Part B: Data Cleaning & Preparation

6. Handle missing values appropriately and explain your method in comments.

```python
# 1. Load dataset
df = pd.read_csv('/content/batch 7 dataset-salary.csv')

# 2. Check missing values
print("Before Cleaning:")
print(df.isnull().sum())

# 3. Handle missing values
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])
df['Job Title'] = df['Job Title'].fillna(df['Job Title'].mode()[0])
df['Company Name'] = df['Company Name'].fillna(df['Company Name'].mode()[0])

# 4. Verify
print("After Cleaning:")
print(df.isnull().sum())
```

7.Check for duplicate records and remove them if present.

```python
print("Before removing duplicates:")
print(df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Verify again
print("After removing duplicates:")
print(df.duplicated().sum())
```

8.Convert relevant columns into correct data types (e.g., salary, experience).

```python

print("Before Conversion:")
print(df.dtypes)

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

df['Salaries Reported'] = pd.to_numeric(df['Salaries Reported'], errors='coerce')
df['Salaries Reported'] = df['Salaries Reported'].fillna(0)
df['Salaries Reported'] = df['Salaries Reported'].astype(int)

df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

print("\nAfter Conversion:")
print(df.dtypes)
```

```python
print(df.dtypes)
```

9.Create a new column categorizing experience levels (Beginner, Intermediate, Expert).

```python

def exp_level(x):
    if x <= 2:
        return "Beginner"
    elif x <= 5:
        return "Intermediate"
    else:
        return "Expert"

df['Experience Level'] = df['Salaries Reported'].apply(exp_level)

print(df[['Salaries Reported', 'Experience Level']].head())
```

10.Standardize or clean text columns if required (e.g., job roles, locations).


```python

# Convert text to lowercase
df['Job Title'] = df['Job Title'].str.lower()
df['Location'] = df['Location'].str.lower()
df['Company Name'] = df['Company Name'].str.lower()

# Remove extra spaces
df['Job Title'] = df['Job Title'].str.strip()
df['Location'] = df['Location'].str.strip()
df['Company Name'] = df['Company Name'].str.strip()

# Display result
print(df[['Job Title', 'Location', 'Company Name']].head())
```

18.Identify any outliers in salary distribution

```python

# Calculate Q1 and Q3
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define lower and upper limits
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]

print("Outliers in Salary:")
print(outliers[['Job Title', 'Salary']])
```

23.Compare NumPy results with Pandas results.


```python
import numpy as np

# NumPy calculations
salary_array = df['Salary'].to_numpy()

np_mean = np.mean(salary_array)
np_std = np.std(salary_array)
np_min = np.min(salary_array)
np_max = np.max(salary_array)

# Pandas calculations
pd_mean = df['Salary'].mean()
pd_std = df['Salary'].std()
pd_min = df['Salary'].min()
pd_max = df['Salary'].max()

# Print comparison
print("NumPy vs Pandas Comparison:\n")

print("Mean -> NumPy:", np_mean, "| Pandas:", pd_mean)
print("Std Dev -> NumPy:", np_std, "| Pandas:", pd_std)
print("Min -> NumPy:", np_min, "| Pandas:", pd_min)
print("Max -> NumPy:", np_max, "| Pandas:", pd_max)
```
